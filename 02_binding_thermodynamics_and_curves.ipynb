## 1. Fraction bound vs. Kd

Sweeps the dissociation constant over several orders of magnitude and solves the 1:1 equilibrium to plot fraction-bound curves for different total-concentration regimes.

<sub>Source script: <code>fraction_bound_vs_kd.ipynb</code></sub>

```python
import numpy as np
import pandas as pd

A_tot = 5.0      # mM
B_tot = 50.0     # mM

# sweep Kd from 100 M down to 1 microM (1 M = 1000 mM, 1 microM = 10^-3 mM)
Kd_vals_mM = np.logspace(5, -3, 50)

fractions = []

for Kd in Kd_vals_mM:
    # solve for AB using the correct formula:
    # Kd = (A_free * B_free) / AB
    # AB satisfies:
    # AB^2 - (A_tot + B_tot + Kd) AB + A_tot * B_tot = 0

    a = 1
    b = -(A_tot + B_tot + Kd)
    c = A_tot * B_tot

    roots = np.roots([a, b, c])

    # choose physically correct root:
    AB = max(r.real for r in roots if np.isreal(r) and 0 <= r.real <= A_tot)

    fraction_bound = AB / A_tot
    fractions.append(fraction_bound)

df_5_50 = pd.DataFrame({
    "Kd_mM": Kd_vals_mM,
    "Kd_nM": Kd_vals_mM * 1e6,
    "Fraction_bound": fractions
})

print(df_5_50)
```

Calculations for 2 A + B -> A2B euqation (2:1 binding coefficient)

```python
import numpy as np
import pandas as pd

A_tot = 10.0   # mM
B_tot = 50.0  # mM

# fractions of A bound (cannot exceed 1.0)
fraction_vals = np.linspace(0.001, 0.999, 50)

Kd_vals = []

for f in fraction_vals:
    Kd = (2 * A_tot * (1 - f)**2 * (B_tot - (f * A_tot) / 2)) / f
    Kd_vals.append(Kd)

df_fraction_to_Kd = pd.DataFrame({
    "Fraction_bound": fraction_vals,
    "Kd_mM": Kd_vals,
    "Kd_nM": np.array(Kd_vals) * 1e6
})

print(df_fraction_to_Kd)
```

```python
import numpy as np
import pandas as pd

A_tot = 10.0   # mM
B_tot = 50.0  # mM

# sweep fraction bound from 0 to 1 (exclusive)
fraction_vals = np.linspace(0.001, 0.999, 50)

Kd_vals = []

for f in fraction_vals:
    B_free = B_tot - 2 * f * A_tot
    if B_free <= 0:
        Kd = np.nan  # physically impossible
    else:
        Kd = ((1 - f) * B_free**2) / f
    Kd_vals.append(Kd)

df_A_B2 = pd.DataFrame({
    "Fraction_bound": fraction_vals,
    "Kd_mM": Kd_vals,
    "Kd_nM": np.array(Kd_vals) * 1e6
})

print(df_A_B2)
```

```python
import numpy as np
import pybindingcurve as pbc
import matplotlib.pyplot as plt

# --- 1. Define the system ---
# "1:1" means a simple A + B ⇌ AB binding equilibrium
system = pbc.BindingCurve("1:1")

# --- 2. Fixed initial concentrations (in n M, for example) ---
A0 = 10.0*10**6  # initial concentration of A
B0 = 50.0*10**6  # initial concentration of B

# --- 3. Variable dissociation constants (Kd) ---
Kd_values = np.logspace(-2, 2, 100)  # from 0.01 to 100 nM

# --- 4. Compute equilibrium [AB] for each Kd ---
AB_eq = []
for kd in Kd_values:
    params = {"p": A0, "l": B0, "kdpl": kd}  # p=protein(A), l=ligand(B)
    y = system.query(params)  # returns fraction bound by default
    # Convert fraction bound to absolute [AB]
    AB_eq.append(y * A0)  # or multiply by limiting species (A0)

AB_eq = np.array(AB_eq)
#print(AB_eq)
# --- 5. Plot results ---
plt.figure()
plt.semilogx(Kd_values, AB_eq, '-')
plt.xlabel("Dissociation constant Kd (nM)")
plt.ylabel("Equilibrium [AB] (nM)")
plt.title("Dependence of [AB] on Kd")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

```python
import numpy as np
import pybindingcurve as pbc
import matplotlib.pyplot as plt

# --- 1. Define the system ---
# "1:1" means a simple A + B ⇌ AB binding equilibrium
system = pbc.BindingCurve("1:1")

# --- 2. Fixed initial concentrations (in n M, for example) ---
A0 = 10.0*10**6  # initial concentration of A
B0 = 50.0*10**6  # initial concentration of B

# --- 3. Variable dissociation constants (Kd) ---
Kd_values = np.logspace(-2, 2, 100)  # from 0.01 to 100 nM

# --- 4. Compute equilibrium [AB] for each Kd and calculate fraction bound ---
fraction_bound_A = []
for kd in Kd_values:
    params = {"p": A0, "l": B0, "kdpl": kd}  # p=protein(A), l=ligand(B)
    fraction_ab = system.query(params)  # returns fraction bound by default
    # Convert fraction bound of A to absolute [AB]
    AB_eq = fraction_ab * A0
    # Calculate fraction of A bound
    fraction_bound_A.append((A0 - AB_eq) / A0)

fraction_bound_A = np.array(fraction_bound_A)

# --- 5. Plot results ---
plt.figure()
plt.semilogx(Kd_values, fraction_bound_A, '-')
plt.xlabel("Dissociation constant Kd (mM)")
plt.ylabel("Fraction of AA Bound")
plt.title("Fraction of AA Bound vs Kd")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```
